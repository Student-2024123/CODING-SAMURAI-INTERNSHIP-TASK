```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# For better visuals
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Fill Age with median
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop Cabin column (too many missing values)
df.drop(columns=['Cabin'], inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='Sex', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
plt.hist(df['Age'], bins=30)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival vs Gender")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title("Survival vs Passenger Class")
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Survival vs Age")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
corr
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Key Insights:
- Females had a much higher survival rate than males
- Passengers in 1st class survived more than those in 3rd class
- Younger passengers had a higher chance of survival
- Fare and Survival show a positive correlation
